## Cleaning outliers for analysis

To address the outliers of this dataset, we will analyze the statistical distribution, plot boxplots and calculate the limits using the IQR Score rule.

Concept: First, let's remember what the IQR is and how it will be applied;

The IQR is calculated by subtracting the Third Quartile (75%) from the First Quartile (25%):

# IQR = Q3 - Q1

## Let's take a look at our data and see what we've identified.


```python
#identifying outliers for the price variable
q1_price = df_clean.price.quantile(.25)
q3_price = df_clean.price.quantile(.75)
IQR_price = q3_price - q1_price
print('Python print for me:''\n'),
print('Interquartile range dimensions IQR:',
      '\n\nIQR = VALIABLES PRICE MEAN: ', IQR_price)

# definindo os limites                                       
sup_price = q3_price + 1.5 * IQR_price
inf_price = q1_price - 1.5 * IQR_price

print('price limit Upper : ', sup_price)
print('price limit Lower: ', inf_price)
```
